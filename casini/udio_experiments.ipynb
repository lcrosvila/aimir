```python
import json
import os
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("notebook")
```

```python
from multiprocessing import Pool
from tqdm.auto import tqdm

def json2df(json_file):
    return pd.read_json('../udio/metadata/' + json_file, orient='index').T

filename = '../udio/metadata.pkl'
try:
    print('opening udio_df from pickle')
    udio_df = pd.read_pickle(filename)
    print(udio_df.shape)
except:
    print('creating udio_df from json files')
    udio_metadata_files = os.listdir('../udio/metadata/')
    print(f'metadata files: {len(udio_metadata_files)}')

    # udio_dfs = [pd.read_json('../udio/metadata/' + f, orient='index').T for f in udio_metadata_files[:]]
    # read json files with multiprocessing and store results in udio_dfs
    # use tqdm to show progress bar
    with Pool(16) as p:
        udio_dfs = list(tqdm(p.imap(json2df, udio_metadata_files), total=len(udio_metadata_files)))
    
    udio_df = pd.concat(udio_dfs).reset_index(drop=True)
    udio_df.to_pickle(filename)

udio_df
```

## language detection

```python
import fasttext
from huggingface_hub import hf_hub_download

model_path = hf_hub_download(repo_id="facebook/fasttext-language-identification", filename="model.bin")
model = fasttext.load_model(model_path)
# model.predict(text)[0]

def detect_language(prompt, model):
    if prompt:
        return model.predict(prompt.replace('\n', ' '))[0][0].replace('__label__','')
    else:
        return None

udio_df['language'] = udio_df['lyrics'].apply(lambda x: detect_language(x, model))
udio_df.to_pickle('../udio/metadata.pkl')
```

```python
# barplot with languages value counts
import matplotlib.pyplot as plt
udio_df.language.value_counts()[:25].plot(
    kind='bar',
    figsize=(15, 5),
    title='Top25 Lyrics Languages (Udio)',
    color='#f77'
    )
plt.show()
```

```python
# select rows in udio_df that are not null and where language startswith eng
df = udio_df.dropna(subset='language')[['lyrics','language']]

df = df[df.language.str.startswith('eng')]
df
```

```python
from transformers import pipeline
classifier = pipeline(
    'zero-shot-classification', 
    model='facebook/bart-large-mnli',
    #model='roberta-large-mnli', 
    # model='microsoft/deberta-xlarge-mnli',
    )

sequence_to_classify = df.iloc[0].lyrics

candidate_labels = ['parody','serious']
hypothesis_template = "These are the lyrics of a {} song."
res = classifier(
    sequence_to_classify, 
    candidate_labels, 
    # multi_label=True, # if more than one choice can be correct
    hypothesis_template=hypothesis_template
    )
# print(res)

for i in enumerate(res['labels']):
    print(f"{res['labels'][i[0]]}: {res['scores'][i[0]]}")

print(f'LYRICS:\n\n{sequence_to_classify}\n\n')
```

## look for artist name in the prompts

```python
# !pip install lxml
import pandas as pd
```

```python
# requires lxml
bsa = pd.read_html('https://en.wikipedia.org/wiki/List_of_best-selling_music_artists')
bsa = pd.concat(bsa[:6]).reset_index(drop=True)[['Artist']]
bsa
```

```python
df = pd.read_pickle('../udio/metadata.pkl')
print(df.columns)
df = df[['lyrics','prompt','tags','replaced_tags','language']]

def check_artist(row, artist_name, column):
    col = row[column].lower()
    if artist_name.lower() in col:
        return True
    else:
        return False

bsa['udio_prompts'] = bsa['Artist'].apply(lambda x: df.apply(lambda row: check_artist(row, x, 'prompt'), axis=1).sum())
bsa['udio_lyrics'] = bsa['Artist'].apply(lambda x: df.apply(lambda row: check_artist(row, x, 'lyrics'), axis=1).sum())
```

```python
bsa['sum'] = bsa['udio_prompts'] + bsa['udio_lyrics']
bsa.sort_values('sum', ascending=False).head(50)
```
